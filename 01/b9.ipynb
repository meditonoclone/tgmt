```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
my_path = "C:\\Users\\ADMIN\\Desktop\\TranTien\\ThiGiacMayTinh\\01\\2.jpg"
im = Image.open(my_path)
plt.imshow(im)
```

```python
gray_img = im.convert('L')
plt.imshow(gray_img)
```

```python
plt.imshow(gray_img, cmap='gray')
```

```python
arr = np.array(gray_img)
print(arr)
```

```python
plt.figure()
plt.hist(arr.flatten(), bins=256)
plt.title('Histogram ca nh xm')
plt.xlabel('Gi tr pixel')
plt.ylabel('S lng')
plt.show()
```
