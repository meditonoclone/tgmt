```python
def display(img):
    cv2.imshow('title', img)
    cv2.waitKey(0)
    cv2.destroyWindow('title')
```

```python
display(img)

```

```python
img2 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
display(img2)
```

```python
display(img)
display(img2)
```

```python
def displayAll(list):
    for key, img in enumerate(list):
        cv2.imshow(f'image {key}', img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
displayAll([img, img2])
```
